```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def data_gen(N, n_miss):
  x =  np.random.randn(N)
  y = (np.sign(x)+1)/2
  x = np.sort(x)
  y = np.sort(y)
  if n_miss>0:
    y[0:n_miss] = 1
    y[-1-n_miss:] = 0
  return x, y

class LogisticClassifier(nn.Module):  
  """
  Subclass nn.Module to implement the logistic regressor as a two layer, two neuron network with the input layer being activated linearly 
  and the output activated via softmax. 
  """

    def __init__(self):
        super(LogisticClassifier, self).__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, vec): 
        return F.sigmoid(self.linear(vec))

def train(x, y, lr=0.1, epochs=1000):
  """
  Wrapper for the training sequence to handle part (3)
  """
  N = len(x)
  x_tensor = Variable(torch.Tensor(x.reshape(N,1)))
  y_tensor = Variable(torch.Tensor(y.reshape(N,1)))
  model = LogisticClassifier()
  optimizer = torch.optim.SGD(model.parameters(), lr=lr) # Optimize using stochastic gradient descent
  cost = nn.BCELoss(size_average=True) # Using binary cross entropy for loss
  loss_vec = [] 
  for epoch in range(epochs):
    optimizer.zero_grad()
    y_pred = model(x_tensor)
    loss = cost(y_pred, y_tensor)
    loss_vec.append(loss.item())
    loss.backward()
    optimizer.step()
  return model, loss_vec

def get_weights(model):
  """
  Given a 1d logistic regressor, returns the weight vector that is orthogonal to the decision boundary. The regressor is of the form sigmoid(mx+b).
  """
  w = list(model.parameters())
  m = w[0].data.numpy()[0][0]
  b = w[1].data.numpy()[0]
  return m, b
```

## (1) No missclassifications in data

```
x, y = data_gen(200, 0)
model, loss_vec = train(x, y)
plt.plot(loss_vec)
```

```
m, b = get_weights(model)
x_ = np.linspace(-2, 2, 200)
plt.plot(sigmoid(b+m*x_))
plt.plot(np.sort(y, axis=0), '.')
print("Transition width: "+str(round(1/m, 4)))
print()
```

For the given paramters, we find that the transition width is $0.188$. Of course, with no data errors, we expect this transition width to vanish with increasing $N$. 

## (2) 5 misses on either side of origin

```
x, y = data_gen(200, 5)
plt.figure()
plt.title("Labels with 5 misses on either side of 0")
plt.stem(y)

model, loss_vec = train(x, y)
m, b = get_weights(model)
x_ = np.linspace(-2, 2, 200)
plt.figure()
plt.plot(sigmoid(b+m*x_))
plt.plot(np.sort(y, axis=0), '.')
print()
print("Transition width: "+str(round(1/m, 4)))
print()
```

As one might expect, with missclassifications in the training set, the transition width increases to 0.6 with the parameters chosen (and depending on the seed obviously). We see that the classification scheme remains robust, at least with a total of 10 (unbiased) errors. 

## (3) $n$ misses on either side of origin

```
n_misses = [15, 20, 25, 30, 35]
width = []
for n_miss in n_misses:
  x, y = data_gen(200, n_miss)
  model, _ = train(x, y)
  m, b = get_weights(model)
  width.append(abs(1/m))
```

```
plt.stem(n_misses, width)
plt.ylabel(r"$\frac{1}{m}$")
```

What's happening here is not surprising. Beyond 25 misses on either side, we actually begin correcting the mislabelings. Let us repeat this on a finer grid to see the trend between the number of misses in the labeling and the transition  width.

```
n_misses = np.arange(1, 50, 2)
width = []
for n_miss in n_misses:
  x, y = data_gen(200, n_miss)
  model, _ = train(x, y)
  m, b = get_weights(model)
  width.append(abs(1/m))
```

```
plt.plot(n_misses, np.log(width), '.--')
plt.ylabel(r"log$\frac{1}{m}$")
```
